```python
import pymorphy2
from nltk.corpus import stopwords
morph = pymorphy2.MorphAnalyzer()
punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~—«»…'
bad_tags = ['PREP', 'CONJ', 'PRCL', 'INTJ', 'NPRO']
stop_words = stopwords.words("russian")
```

```python
def file_to_words(file_path):
    file = open(file_path, 'r', encoding='utf-8')
    text = file.read().split('\n')
    file.close()
    words = ''
    for line in text:
        if len(line) == 0 or line[0] == '[':
            continue
        for word in line.split():
            word = word.lstrip(punctuation)
            word = word.rstrip(punctuation)
            if len(word) == 0:
                continue
            parsed = morph.parse(word)[0]
            if parsed.tag.POS in bad_tags or parsed.normal_form in stop_words:
                continue
            words += parsed.normal_form + ' '
    return words
    
# file_to_vec('data/Aikko/10 негритят (10 African Americans).txt')
```
